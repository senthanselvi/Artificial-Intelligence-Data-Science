```python
#GENETIC
import random
 
def fitness(solution):
    non_attacking = 0
    for i in range(len(solution)):
        for j in range(i + 1, len(solution)):
            if solution[i] != solution[j] and abs(solution[i] - solution[j]) != j - i:
                non_attacking += 1
    return non_attacking

 
def crossover(parent1, parent2):
    point = len(parent1) // 2
    return parent1[:point] + parent2[point:], parent2[:point] + parent1[point:]

 
def mutate(solution):
   i, j = random.sample(range(len(solution)), 2) 
   solution[i], solution[j] = solution[j], solution[i]

 
def genetic_algorithm(n, board, generations):
    population = [board]
    for gen in range(generations):
        parent1, parent2 = population[0], population[0]   

  
        offspring1, offspring2 = crossover(parent1, parent2)

         
        if gen % 2 == 0: mutate(offspring1)
        if gen % 3 == 0: mutate(offspring2)

        population = [offspring1, offspring2]

        
        best_solution = max(population, key=fitness)
        if fitness(best_solution) == n * (n - 1) // 2:
            print(best_solution)
            return best_solution, gen + 1   

    return population[0], generations  

 
n = 8
board = [0, 4, 7, 5, 2, 6, 1, 3]  

 
generations = 1000

 
solution, gen = genetic_algorithm(n, board, generations)
print(f"Solution at generation {gen}: {solution}")
print(f"Fitness (non-attacking pairs): {fitness(solution)}")
```
