```python
#BFS
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt

#class
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.explored = []  # To track explored nodes
        self.frontier = []  # To maintain the BFS frontier
    
    def addEdge(self, u, v):
        self.graph[u].append(v)
    
    def BFS(self, initial_state, goal_state):
        self.explored.append(initial_state)
        self.frontier.append(initial_state)
        path_str = ''
        
        while self.frontier:
            if goal_state == self.frontier[0]:
                element = self.frontier.pop(0)
                path_str += element
                return path_str
            else:
                element = self.frontier.pop(0)
                path_str += element

            for adj in self.graph[element]:
                if adj not in self.explored:
                    self.explored.append(adj)
                    self.frontier.append(adj)
    
    
    def disp(self):
        print(self.graph)

if __name__ == '__main__':
    g = Graph()
    g.addEdge('a', 'b')
    g.addEdge('a', 'c')
    g.addEdge('b', 'd')
    g.addEdge('b', 'e')
    g.addEdge('d', 'g')
    g.addEdge('c', 'f')
    g.addEdge('f', 'h')

    start_node = 'a'
    goal_node = 'h'
    path = g.BFS(start_node, goal_node)
    print(path)
        
    
```

```python
#DFS
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt

#class
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.explored = []  # To track explored nodes
        self.frontier = []  # To maintain the BFS frontier
    
    def addEdge(self, u, v):
        self.graph[u].append(v)
    
    def DFS(self, initial_state, goal_state):
        self.explored.append(initial_state)
        self.frontier.append(initial_state)
        path_str = ''
        
        while self.frontier:
            if goal_state == self.frontier[0]:
                element = self.frontier.pop()
                path_str += element
                return path_str
            else:
                element = self.frontier.pop()
                path_str += element

            for adj in reversed(self.graph[element]):
                if adj not in self.explored:
                    self.explored.append(adj)
                    self.frontier.append(adj)
    
    
    def disp(self):
        print(self.graph)

if __name__ == '__main__':
    g = Graph()
    g.addEdge('a', 'b')
    g.addEdge('a', 'c')
    g.addEdge('b', 'd')
    g.addEdge('b', 'e')
    g.addEdge('d', 'g')
    g.addEdge('c', 'f')
    g.addEdge('f', 'h')

    start_node = 'a'
    goal_node = 'h'
    path = g.DFS(start_node, goal_node)
    print(path)
        
    
```

```python
#UCS
import heapq

graph = {
    'S': {'A': 3, 'B': 1, 'C': 8},
    'A': {'S': 3, 'D': 3, 'E': 7, 'G': 15},
    'B': {'S': 1, 'G': 20},
    'C': {'G': 5, 'S': 8},
    'D': {},
    'E': {},
    'G': {}
}

def ucs(graph, start, goal):
    queue = [(0, start)]
    cost = {start: 0}
    parent = {start: None}

    while queue:
        m, n = heapq.heappop(queue)
        
        if n == goal:
            break
        
        for i, w in graph[n].items():
            nc = m + w
            
            if i not in cost or nc < cost[i]:
                cost[i] = nc
                heapq.heappush(queue, (nc, i))
                parent[i] = n

    path = []
    node = goal
    while node is not None:
        path.append(node)
        node = parent[node]
    path.reverse()
    
    return path, cost[goal]

# Print the result of UCS from 'S' to 'G'
print(ucs(graph, 'S', 'G'))
```

```python
#DLS
class dls:
    def __init__(self):
        self.graph = {}
        
    def addEdge(self, u , v):
        self.graph[u] = self.graph.get(u, []) + [v]
        self.graph[v] = self.graph.get(v, []) + [u]
        
    def dls(self, start, goal, limit):
        def dls_rec(node, goal, path, visited, limit):
            if node == goal:
                return path
            if limit <=0 :
                return None
            visited.append(node)
            for adj in self.graph[node]:
                if adj not in visited:
                    new_path = dls_rec(adj, goal, path + [adj], visited, limit-1)
                    if new_path:
                        return new_path
            return None
        visited = []
        path = dls_rec(start, goal, [start], visited, limit)
        return path 

    def disp(self):
        print(self.graph)

g = dls()
g.addEdge('a', 'b')
g.addEdge('a', 'c')
g.addEdge('b', 'd')
g.addEdge('b', 'e')
g.addEdge('d', 'g')
g.addEdge('c', 'f')
g.addEdge('f', 'h')    

path = g.dls('a','h',3)
print(path)
```
