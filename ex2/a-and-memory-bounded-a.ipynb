```python
#A*
def heuristic(node):
    heuristic_values = {
        "A": 366,
        "B": 374,
        "C": 329,
        "D": 244,
        "E": 253,
        "F": 178,
        "G": 193,
        "H": 98,
        "I": 0
    }
    return heuristic_values[node]

graph = {
    "A": [("B", 75), ("C", 118), ("E", 140)],
    "B": [("A", 75)],
    "C": [("A", 118), ("D", 111)],
    "D": [("C", 111)],
    "E": [("A", 140), ("G", 80), ("F", 99)],
    "F": [("E", 99), ("I", 211)],
    "G": [("E", 80), ("H", 97)],
    "H": [("G", 97), ("I", 101)],
    "I": [("H", 101), ("F", 211)]
}

def astar_search(initial, goal):
    visited = []
    frontier = [(0+heuristic(initial),0,initial,[])]
    while frontier:
        frontier.sort(key = lambda x : x[0])
        if frontier[0][2] == goal:
            f, g, element, path = frontier.pop(0)
            return path + [element], f
        else:
            f, g, element, path = frontier.pop(0)
            visited.append(element)
            for adj,weight in graph[element]:
                if adj not in visited:
                    new_g = g + weight
                    new_f = new_g + heuristic(adj)
                    new_path = path + [element]
                    frontier.append((new_f, new_g, adj, new_path))

path, cost = astar_search('A','I')
print(path, cost)
```
